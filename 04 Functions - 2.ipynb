### Library Modules
- A library module is a file with the extension .py containing functions and variable that can be used (we say imported) by any program  
    -- The library module can be created in IDLE or any text editor and looks like an ordinary Python program  

- To gain access to the functions and variables of a library module, place a statement of the form __import mouldeName__ at the beginning of the program  
    -- Any function from the module can be used in the program by prepending the function name with the module name followed by a period

Assuming that the function pay in the example program on p.14 is contained in a file named finance.py that is located in the same folder as the example, the example could be rewritten as

```python
def main():
    ##Calculate a person's weelky pay.
    hourlyWage = float(input("Enter houly wage: "))
    hoursWorked = int(input("Enter # hours worked: "))
    earnings = pay(hourlyWage, hoursWorked)
    print(earnings)

def pay(wage, hours):
    if hours <= 40:
        amount = wage * hours
    else:
        amount = (wage * 40) + ((1.5) * wage * (hours - 40))
    return amount

main()
```

### List Comprehension
- if list1 is a list, then the following statement creates a new list, list2, and places f(item) into the list for each item in list1  
    list2 = [f(x) for x in list1] # list1 안에 있는 x 하나하나에 함수 f를 적용시키는 것이다.  
    where f is either a Python built-in function or a user-defined function

```python
list1 = ['1', '23', '-5', '17']
list2 = [int(x) for x in list1]
print(list2)
list3 = [x * 2 for x in list2]
print(list3)
def minus(x):
    res = x * (-1)
    return res
list4 = [minus(x) for x in list2]
print(list4)
```

- The for clause in a list comprehension can optionally be followed by an if clause

```python
[minus(x) for x in list2 if x >= 0]
```

- List comprehension can be applied to objects other thans lists, such as, strings, tuples, and arithmetic progressions generated by range functions

```python
[ord(x) for x in 'abc']  # ord() operator는 표준에 따라 코드값을 반환해준다. 십진법화를 시켜준다.
```

```python
[x ** .5 for x in (-1 , 5, 2, -23) if x >= 0]
```

### Default Argument Values
- A typical format for a function definition using default values is   
    def functionName(par1, par2, par3=value3, par4=value4) 이며 디폴트 값이 없으면 argument가 반드시 주어저야 한다. 

```python
def main():
    say('Hi')
    say('Hi', 5)
def say(par1, par2=1):
    print(par1 * par2)
main()
```

### Keyword Arguments
- Arguments can be passed to functions by using the names of the corresponding parameters instead of relying on position

```python
def main():
    find(5, 10)
    find(x=7, z=15)
    find(y=8, z=-12)
def find(x=1, y=2, z=3):
    print('x is', x, ',y is', y, 'and z is', z)
main()
```

### Lambda Expression
- Lambda expressions are one-line mini-functions  
    __lambda par1, par2, ...: expression__

```python
names = ['Dennis Ritchie', 'Alan Kay', "John Backus", "James Gosling"]
names.sort(key=lambda name: name.split()[-1])  # key에 쓰는 함수를 한줄로 간단하게 표현
print(names)
nameString = ", ".join(names)  # 이러면 깔끔하게 출력이 가능하다.
print(nameString)
```
